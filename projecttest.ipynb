Now, we are going to plot any single independent variable against our dependent

```python
df= dataset 
```

```python
plt.scatter(x = 'Temperature', y = 'Stress', data = df, s = 100, alpha = 0.3, edgecolor = 'white')
plt.title('Temperature Vs Stress', fontsize = 16)
plt.ylabel('Stress', fontsize = 12)
plt.xlabel('Temperature', fontsize = 12)

plt.savefig('temperature.png')
```

```python
X_var = df[['temperature']] # independent variable
y_var = df['stress'] # dependent variable
```

using statmodel

```python
slr_model = sm.OLS(y_var, X_var) # Ordinary Least Squares 
slr_reg = slr_model.fit()
```
